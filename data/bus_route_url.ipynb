```python
import pandas as pd
```

```python
df = pd.read_csv('./Bus_route_url.csv')
display(df.head())
```

```python
type_to_chinese = {
    'Green': '綠',
    'Blue': '藍',
    'Brown': '棕',
    'Orange': '橘',
    'Yellow':'黃',
    'Red':'紅',
    'Urban':'市區'
}
df['TypeZh'] = df['Type'].map(type_to_chinese)
display(df.head())
```

```python
# temp = df.assign(RouteName=df['RouteID'].replace(['G', 'B', 'O', 'Y', 'R', 'UR'], ['綠', '藍', '橘', '黃', '紅', '市區'], regex=True))
# Define the replacement patterns and their corresponding replacements
patterns = [r'^(G)(\d.*)$', r'^(B)(\d.*)$', r'^(BR)(\d.*)$', r'^(O)(\d.*)$', r'^(Y)(\d.*)$', r'^(R)(\d.*)$', r'^(UR)(\d.*)$']
replacements = ['綠\\2', '藍\\2', '棕\\2', '橘\\2', '黃\\2', '紅\\2', '市區\\2']

# Perform the replacements
df['RouteName'] = df['RouteID'].replace(patterns, replacements, regex=True)
display(df)
```

```python
certain_route = {
    'Green': '綠幹線',
    'Blue': '藍幹線',
    'Brown': '棕幹線',
    'Orange': '橘幹線',
    'Yellow': '黃幹線',
    'Red': '紅幹線',
    'Urban': '市區'
}
df['RouteName'] = df.apply(lambda row: certain_route[row['Type']] if row['RouteName'] == row['Type'] else row['RouteName'], axis=1)
display(df)
```

```python
df.to_csv('bus.csv', encoding='UTF-8', index=False)
```
